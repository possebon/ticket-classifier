```python
import pandas as pd
import matplotlib.pyplot as plt
import json

with open('data/preprocess/tickets_labeled.json') as f:  
    data = json.load(f)
    df=pd.DataFrame(data)
```

```python
cat_totals = df.groupby("service_name")["ticket_number"].count().sort_values()

```

```python
cat_totals
cat_totals.plot(kind="barh", fontsize=4)
```
